```python
breach['Reported_Year'] = pd.DatetimeIndex(breach['Date Reported To OCA']).year
```

```python
breach.head()
```

```python
breach.columns = breach.columns.str.replace(' ', '_')
```

```python
breach.columns
```

```python
breach.head()
```

```python
# Looking for null values
breach.isnull().sum()
```

```python
# Replacing null values with 'No'
breach.iloc[:, 7:] = breach.iloc[:, 7:].fillna('No')
```

```python
breach['Breach_Occur_at_the_Reporting_Entity?'] = breach['Breach_Occur_at_the_Reporting_Entity?'].fillna('No')
```

```python
breach.isnull().sum()
```

```python
# Are there different types of Breach?
breach.Breach_Type_Description.unique()
```

```python
# Filter considering 'Electronic' and 'Both' breach types
breach = breach[(breach.Breach_Type_Description == 'Electronic') |  (breach.Breach_Type_Description == 'Both')] 
breach
```

```python
# Still having null values? 
breach.isnull().sum()
```

```python
# Let's drop those 75 rows
breach = breach.dropna(subset=['MA_Residents_Affected'])
breach.isnull().sum()
```

```python
#As we have dropped some rows, let's reset the index and store in a new variable
clean_br = breach.reset_index(drop=True)
clean_br
```

```python
plt.figure(figsize=(7,5))
ax=sns.countplot(x=clean_br.Breach_Type_Description)
for container in ax.containers:
    ax.bar_label(container)
plt.xlabel('Breach Type Description')
plt.title('Breach Type')
plt.show()
```

```python
plt.figure(figsize=(7,5))
groupedvalues = clean_br.groupby('Reported_Year').count().reset_index()
ax = sns.barplot(x='Reported_Year', y='MA_Residents_Affected',
                 data=groupedvalues,
                 errwidth=0)
ax.bar_label(ax.containers[0])
plt.xlabel('Reported Year')
plt.ylabel('Residents Affected')
plt.title('Residents Affected per year')
plt.show()
```
