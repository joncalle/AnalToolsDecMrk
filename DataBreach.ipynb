# **Data Cleaning/Transformation**

---



```python
# Replacing null values with 'No'
breach.iloc[:, 7:] = breach.iloc[:, 7:].fillna('No')
```

```python
# Replacing null values with 'No'
breach['Breach_Occur_at_the_Reporting_Entity?'] = breach['Breach_Occur_at_the_Reporting_Entity?'].fillna('No')
```

```python
# Which columns still have null values?
breach.isnull().sum()
```

```python
# Are there different types of Breach?
breach.Breach_Type_Description.unique()
```

```python
# Filter considering 'Electronic' and 'Both' breach types
breach = breach[(breach.Breach_Type_Description == 'Electronic') |  (breach.Breach_Type_Description == 'Both')] 
breach
```

```python
# Still having null values? 
breach.isnull().sum()
```

```python
# Let's drop those 75 rows because we don't have information to help fill these values
breach = breach.dropna(subset=['MA_Residents_Affected'])
breach.isnull().sum()
```

```python
#As we have dropped some rows, let's reset the index and store in a new variable
clean_br = breach.reset_index(drop=True)
clean_br
```

```python
# Visualizing the breach type distribution after filter those we are going to use in our project
plt.figure(figsize=(7,5))
ax=sns.countplot(x=clean_br.Breach_Type_Description)
for container in ax.containers:
    ax.bar_label(container)
plt.xlabel('Breach Type Description')
plt.title('Breach Type')
plt.show()
```

```python
#Looking for patterns or relationship between the data
corr_mat=clean_br.corr(method='pearson')
sns.set_style('whitegrid')
mask=np.zeros_like(corr_mat,dtype=bool)
mask[np.triu_indices_from(mask)]=True
fig,ax=plt.subplots(figsize=(6,6))
cmap=sns.diverging_palette(226,15,as_cmap=True)
sns.heatmap(corr_mat,mask=mask,cmap=cmap,vmin=-1,vmax=1,center=0,square=True,linewidth=0.5,cbar_kws={'shrink':1},annot=True,annot_kws={'size':15})
plt.tight_layout()
fig.show()
```

```python
# Answer 1 - Top 3 Organizations more Ocurrences and their trend over the years?
br_top_org = pd.DataFrame(clean_br['Organization_Name'].value_counts())
br_top_org.reset_index(level=0, inplace=True)
br_top_org.columns = ['Organization_Name','Count_qtd']
br_top_org_plot = br_top_org.iloc[0:10,:]

ax=sns.barplot(y='Organization_Name', x='Count_qtd', data=br_top_org_plot)
for container in ax.containers:
    ax.bar_label(container)
ax.set_xlim([0, 700])
plt.xlabel("Ocurrences")
plt.ylabel("Organizations")
plt.title('Top 10 Organizations affected by data breaches')
plt.show()
```

```python
# Answer 1 - Top 3 Organizations more Ocurrences and their trend over the years?
# Creating a new dataset with the information we need (selection of columns, group, created a count column, filter for the top 3 organization) and create a plot that is easy to understand
br_top_org_y = pd.DataFrame()
br_top_org_y['Organization_Name'] = clean_br.Organization_Name
br_top_org_y['Reported_Year'] = clean_br.Reported_Year
br_top_org_y['Assigned_Breach_Number'] = clean_br.Assigned_Breach_Number
br_top_org_y_plot = br_top_org_y.groupby(['Organization_Name','Reported_Year'])['Assigned_Breach_Number'].count()
br_top_org_y_plot = br_top_org_y_plot.to_frame(name='Count_qtd').reset_index()
br_top_org_y_plot = br_top_org_y_plot[(br_top_org_y_plot["Organization_Name"] == "Mutual One Bank" ) + (br_top_org_y_plot["Organization_Name"] == "The Village Bank") + (br_top_org_y_plot["Organization_Name"] == "St. Mary's Credit Union")]

plt.figure(figsize=(7,5))
ax=sns.lineplot(x=br_top_org_y_plot['Reported_Year'],
                y=br_top_org_y_plot['Count_qtd'],
                hue=br_top_org_y_plot['Organization_Name'],
                legend=True)
ax.set_ylim([0, 220])
plt.xlabel("Years")
plt.ylabel("Number of Ocurrences")
plt.title('Top 3 Organizations - Ocurrences over the years')
plt.show()
```

```python
# Answer 2 - What type of data stolen through presents the greatest growth over the years?
# Creating a new dataset with the information we need and also verifying the distinct sentences to transform them and create a plot that is easy to understand
clean_br_plot = clean_br.iloc[:, 7:11]
clean_br_plot['Assigned_Breach_Number'] = clean_br.Assigned_Breach_Number
clean_br_plot['Reported_Year'] = clean_br.Reported_Year
clean_br_plot = clean_br_plot.melt(id_vars=["Assigned_Breach_Number",'Reported_Year'], var_name="Data_breached", value_name="Value").reset_index(drop=True)
clean_br_plot = clean_br_plot[clean_br_plot['Value']=='Yes']
clean_br_plot['Data_breached'].unique()
```

```python
# Answer 2 - What type of data stolen through presents the greatest growth over the years?
# Adjusting the Data breached content so the plot is easily to understand and creating the final plot
clean_br_plot['Data_breached'] = clean_br_plot['Data_breached'].str.replace('SSNBreached','Social Security Number')
clean_br_plot['Data_breached'] = clean_br_plot['Data_breached'].str.replace('Account_Number_Breached','Account Number')
clean_br_plot['Data_breached'] = clean_br_plot['Data_breached'].str.replace('Drivers_Licenses_Breached','Drivers License')
clean_br_plot['Data_breached'] = clean_br_plot['Data_breached'].str.replace('Credit_Debit_Numbers_Breached','Credit/Debit Card')

ax=sns.displot(clean_br_plot, x="Reported_Year", 
            col="Data_breached",
            bins=7,
            discrete=True,
            shrink=0.8)
ax.fig.suptitle('Type of data breached over the years', size=16, 
               y=1.1)
ax.set(xlabel='Year', 
      ylabel='Count')
ax.set_titles('{col_name}')
plt.show()
```

```python
# Answer 2 - What type of data stolen through presents the greatest growth over the years?
# Calculating the increase/decrease over the years for each type of data
clean_br_plot_2 = clean_br_plot[(clean_br_plot['Reported_Year'] == 2016) | (clean_br_plot['Reported_Year'] == 2022)]
clean_br_plot_2_table = clean_br_plot_2.groupby(["Data_breached","Reported_Year"])["Assigned_Breach_Number"].count()
clean_br_plot_2_table = clean_br_plot_2_table.to_frame(name='Coun_qtd').reset_index()
clean_br_plot_2_table_vf = clean_br_plot_2_table.pivot(index='Data_breached', columns='Reported_Year')['Coun_qtd'].reset_index()
clean_br_plot_2_table_vf['%'] = round((clean_br_plot_2_table_vf[2022] / clean_br_plot_2_table_vf[2016] -1 )* 100,2)
clean_br_plot_2_table_vf['diff'] = clean_br_plot_2_table_vf[2022] - clean_br_plot_2_table_vf[2016]
clean_br_plot_2_table_vf
```

```python
# Answer 3 - What type of data stolen through data breaches has the greatest impact on the population of Massachusetts?
clean_br_plot_3 = clean_br.iloc[:, 7:11]
clean_br_plot_3['Assigned_Breach_Number'] = breach.Assigned_Breach_Number
clean_br_plot_3['MA_Residents_Affected'] = breach.MA_Residents_Affected
clean_br_plot_3 = clean_br_plot_3.melt(id_vars=["Assigned_Breach_Number","MA_Residents_Affected"], var_name="Data_breached", value_name="Value").reset_index(drop=True)
clean_br_plot_3 = clean_br_plot_3[clean_br_plot_3['Value']=='Yes']
clean_br_plot_3_vf = clean_br_plot_3.groupby(["Data_breached"])["MA_Residents_Affected"].sum()
clean_br_plot_3_vf = clean_br_plot_3_vf.to_frame(name='sum_qtd').reset_index()

ax=sns.barplot(x=clean_br_plot_3_vf['Data_breached'],
               y=clean_br_plot_3_vf['sum_qtd']/1000 )
for container in ax.containers:
    ax.bar_label(container)
ax.set_xticklabels(["Account number","Credit/Debit Card","Driver's License","Social Security"]) 
plt.xticks(rotation = 45)
ax.set_ylim([0, 6000])
plt.xlabel("Data Breached Type")
plt.ylabel("Number of Residents Affected (x1000)")
plt.title('Residents Affected per type of data stolen')
plt.show()
```

```python
# Answer 3 - What type of data stolen through data breaches has the greatest impact on the population of Massachusetts?
# Calculating the share of number of residents by the data stolen over these years
clean_br_plot_3_vf['Total'] = clean_br_plot_3_vf['sum_qtd'].sum()
clean_br_plot_3_vf['%'] = round((clean_br_plot_3_vf['sum_qtd'] / clean_br_plot_3_vf['Total'] )* 100,2)
clean_br_plot_3_vf
```
