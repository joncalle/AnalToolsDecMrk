```python
df['Reported_Year'] = pd.DatetimeIndex(df['Date Reported To OCA']).year
```

```python
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.head()
```

```python
plt.figure(figsize=(7,5))
ax=sns.countplot(x=df.Breach_Type_Description)
for container in ax.containers:
    ax.bar_label(container)
plt.xlabel('Breach Type Description')
plt.title('Breach Type')
plt.show()
```

```python
plt.figure(figsize=(7,5))
groupedvalues = df.groupby('Reported_Year').count().reset_index()
ax = sns.barplot(x='Reported_Year', y='MA_Residents_Affected',
                 data=groupedvalues,
                 errwidth=0)
ax.bar_label(ax.containers[0])
plt.xlabel('Reported Year')
plt.ylabel('Residents Affected')
plt.title('Residents Affected per year')
plt.show()
```
