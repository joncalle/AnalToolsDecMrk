```python
# Visualizing the breach types distribution
plt.figure(figsize=(7,5))
ax=sns.countplot(x=breach.Breach_Type_Description)
for container in ax.containers:
    ax.bar_label(container)
plt.xlabel('Breach Type')
plt.ylabel('Count')
plt.title('Breach Type')
plt.show()
```

```python
# Checking the maximun value for MA_Residents_Affected so we can use to create ranges
breach.MA_Residents_Affected.max()
```

```python
# Creating a column spliting the number of affected residents by ranges and visualizing the Account Number breach distribution
breach['Range_Num_Afect'] = pd.cut(breach['MA_Residents_Affected'], bins=[0,50,1000,10000,100000,1000000,2982421], labels=['0 to 100','101 to 1k','1k to 10k','10k to 100k','100k to 1M', '>1M'])

plt.figure(figsize=(7,5))
sns.countplot(x=breach.Range_Num_Afect)
for container in ax.containers:
    ax.bar_label(container)
plt.xlabel('Number of Residents Affected')
plt.ylabel('Count')
plt.title('Massachussetts Residents Affected')
plt.show()
```

```python
# Creating a dataframe with the information need for next grouping visualizations (Columns, # of NAs, # of valid values)
breach_na = pd.DataFrame(breach.isna().sum())
breach_na.reset_index(level=0, inplace=True)
breach_na.columns = ['Col_name','NA']
breach_na['YES'] = len(breach) - breach_na['NA']
breach_na = breach_na.melt(id_vars=["Col_name"], var_name="Type_value", value_name="Value").reset_index(drop=True)
breach_na
```

```python
# Filtering the column and visualizing the Social Security Number breach events distribution
breach_na_plot = breach_na.loc[breach_na.Col_name =='SSNBreached']

plt.figure(figsize=(7,5))
ax=sns.barplot(x=breach_na_plot.Type_value, y=breach_na_plot.Value)
for container in ax.containers:
    ax.bar_label(container)
plt.xlabel('SSN Breach?')
plt.ylabel('Number of Occurencies')
plt.title('Data Breaches Ocurrencies - Social Security Number')
plt.show()
```

```python
# Filtering the column and visualizing the Account Number breach events distribution
breach_na_plot = breach_na.loc[breach_na.Col_name =='Account_Number_Breached']

plt.figure(figsize=(7,5))
ax=sns.barplot(x=breach_na_plot.Type_value, y=breach_na_plot.Value)
for container in ax.containers:
    ax.bar_label(container)
plt.xlabel('Account Number Breach?')
plt.ylabel('Number of Occurencies')
plt.title('Data Breaches Ocurrencies - Account Number')
plt.show()
```

```python
# Filtering the column and visualizing the Driver's License Number breach events distribution
breach_na_plot = breach_na.loc[breach_na.Col_name =='Drivers_Licenses_Breached']

plt.figure(figsize=(7,5))
ax=sns.barplot(x=breach_na_plot.Type_value, y=breach_na_plot.Value)
for container in ax.containers:
    ax.bar_label(container)
plt.xlabel('Drivers License Number Breach?')
plt.ylabel('Number of Occurencies')
plt.title('Data Breaches Ocurrencies - Drivers License Number')
plt.show()
```

```python
# Visualizing the Credit/Debit numbers breach distribution
breach_na_plot = breach_na.loc[breach_na.Col_name =='Credit_Debit_Numbers_Breached']

plt.figure(figsize=(7,5))
ax=sns.barplot(x=breach_na_plot.Type_value, y=breach_na_plot.Value)
for container in ax.containers:
    ax.bar_label(container)
plt.xlabel('Credit/Debit Card Breach?')
plt.ylabel('Number of Occurencies')
plt.title('Data Breaches Ocurrencies - Credit/Debit Card')
plt.show()
```

```python
# Visualizing the number of occurencies over the years
plt.figure(figsize=(7,5))
ax = sns.countplot(x=breach.Reported_Year)
for container in ax.containers:
    ax.bar_label(container)
plt.xlabel('Reported Year')
plt.ylabel('Occurencies')
plt.title('Data breaches occurencies over the years')
plt.show()
```

```python
# Visualizing the number of residents affected over the years
plt.figure(figsize=(7,5))
groupedvalues = (breach.groupby('Reported_Year')['MA_Residents_Affected'].sum()/1000).reset_index()
ax = sns.barplot(x='Reported_Year', y='MA_Residents_Affected', data=groupedvalues, errwidth=0)
ax.bar_label(ax.containers[0])
plt.xlabel('Reported Year')
plt.ylabel('Residents Affected (x1000)')
plt.title('Residents Affected over the years')
plt.show()
```

# **Data Cleaning**

---



```python
# Replacing null values with 'No'
breach.iloc[:, 7:] = breach.iloc[:, 7:].fillna('No')
```

```python
# Replacing null values with 'No'
breach['Breach_Occur_at_the_Reporting_Entity?'] = breach['Breach_Occur_at_the_Reporting_Entity?'].fillna('No')
```

```python
# Which columns still have null values?
breach.isnull().sum()
```

```python
# Are there different types of Breach?
breach.Breach_Type_Description.unique()
```

```python
# Filter considering 'Electronic' and 'Both' breach types
breach = breach[(breach.Breach_Type_Description == 'Electronic') |  (breach.Breach_Type_Description == 'Both')] 
breach
```

```python
# Still having null values? 
breach.isnull().sum()
```

```python
# Let's drop those 75 rows because we don't have information to help fill these values
breach = breach.dropna(subset=['MA_Residents_Affected'])
breach.isnull().sum()
```

```python
#As we have dropped some rows, let's reset the index and store in a new variable
clean_br = breach.reset_index(drop=True)
clean_br
```

```python
# Visualizing the breach type distribution after filter those we are going to use in our project
plt.figure(figsize=(7,5))
ax=sns.countplot(x=clean_br.Breach_Type_Description)
for container in ax.containers:
    ax.bar_label(container)
plt.xlabel('Breach Type Description')
plt.title('Breach Type')
plt.show()
```

```python
#Visualizing Possible Outliers - Number of Residents Affected
plt.figure(figsize=(7,5))
sns.boxplot(x=breach['MA_Residents_Affected'], orient = 'h')
plt.xlabel('Residents Affected')
plt.title('Possible outliers on: Number of Residents Affected')
plt.show()
```
